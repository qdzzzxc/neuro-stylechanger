```python
import requests
from lxml import html
import os
import time
from PIL import Image
```

Для обучения CycleGAN для преобразования стилей необходим датасет, в котором присутствуют картины какого стиля, который мы будем копировать, а также обычные фотографии различных вещей. Спарсим нужные картинки для составления датасета.

```python
url = 'https://gallerix.ru/album/Serov'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
}
 
response = requests.get(url, headers=headers)
```

```python
page = html.fromstring(response.text)
paragraphs = page.xpath("//div[@class='pic']/a[@class='animsition-link']/@href")
```

```python
subdirectory = "serov"
os.makedirs(subdirectory, exist_ok=True)

for i, link in enumerate(paragraphs):
    temp_url = 'https://gallerix.ru' + link
    search_for_pic = requests.get(temp_url, headers=headers)
    page = html.fromstring(search_for_pic.text)
    image_url = page.xpath('//p[@class="xpic"]/a/img/@src')[0]
    
    response = requests.get(image_url, headers=headers)
    if response.status_code == 200:
        file_path = os.path.join(subdirectory, f"image_{i}.jpg")
        with open(file_path, "wb") as f:
            f.write(response.content)
        print(f"Картинка {i} успешно скачана")
    else:
        print(f"Ошибка при скачивании картинки. Код состояния: {response.status_code}")
    
    time.sleep(1)
```

Привожу картинки к размеру 256 на 256

```python
current_directory = os.getcwd()
thumbnaildirectory = "serov256x256"
os.makedirs(thumbnaildirectory, exist_ok=True)
new_width, new_height = 256, 256

for image in os.listdir(subdirectory):
    img = Image.open(os.path.join(subdirectory, image))
    w, h = img.size
    if w > h:
        ratio = w/h
        size = (round(ratio * 256), 256)
    else:
        ratio = h/w
        size = (256, round(ratio * 256))
    img = img.resize(size)
    width, height = img.size
    
    left = (width - new_width)/2
    top = (height - new_height)/2
    right = (width + new_width)/2
    bottom = (height + new_height)/2

    img = img.crop((left, top, right, bottom))

    file_path = os.path.join(current_directory, thumbnaildirectory, image)
    
    img.save(file_path)
```
